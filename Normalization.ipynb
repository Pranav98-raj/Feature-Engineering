```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('wines.csv',usecols=[0,1,2])
df.columns=['Class label','Alcohol','Malic Acid']
```

```python
df.head()
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python
sns.scatterplot(df['Malic Acid'],df['Alcohol'],hue=df['Class label'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size=.2)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaler
```

```python
scaler.fit(x_train)
```

```python
x_train_scale=scaler.transform(x_train)
x_test_scale=scaler.transform(x_test)
```

```python
x_train_scale
```

```python
x_train_scale=pd.DataFrame(x_train_scale,columns=x_train.columns)
x_test_scale=pd.DataFrame(x_test_scale,columns=x_test.columns)
```

```python
np.round(x_train_scale.describe(),1)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.scatterplot(x_train['Alcohol'],x_train['Malic Acid'])
plt.subplot(1,2,2)
sns.scatterplot(x_train_scale['Alcohol'],x_train_scale['Malic Acid'])
```
